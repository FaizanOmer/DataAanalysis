# Goal 7: üìä

The "Reviews" column, a numeric field, contains information about the number of reviews users have left when downloading or using the app. This column spans a wide range:

Minimum: 0
Maximum: 22,685,334

To simplify and present this extensive data in a more concise format, we propose creating a new column called "ReviewCategory." Here's how the categorization would look:

If an app has reviews between 0 and 10,000 ‚Üí Categorized as "Less than 10K"
If an app has reviews between 10,000 and 500,000 ‚Üí Categorized as "Between 10K and 500K"
If an app has reviews between 500,000 and 1,000,000 ‚Üí Categorized as "Between 500K and 1Mil"
If an app has reviews between 1,000,000 and 22,685,334 ‚Üí Categorized as "Million Plus"

```python
df['ReviewCategory'] = 'NoReviewsProvided'
df.loc[(df['Reviews'] > 0) & (df['Reviews'] <= 10000.0),'ReviewCategory'] = 'Less than 10K'
df.loc[(df['Reviews'] > 10000) & (df['Reviews'] <= 500000.0),'ReviewCategory'] = 'Between 10K and 500K'
df.loc[(df['Reviews'] > 500000) & (df['Reviews'] <= 1000000),'ReviewCategory'] = 'Between 500K and 1Mil'
df.loc[(df['Reviews'] > 1000000) & (df['Reviews'] <= 22685334),'ReviewCategory'] = 'Million Plus'
df['ReviewCategory'].value_counts()
```
